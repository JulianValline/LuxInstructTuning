```python
import json

# load wiki set
with open("/content/drive/My Drive/Master Thesis/instruction_tuning_wiki.json") as f:
  wiki_data = json.load(f)

# load RTL set
with open("/content/drive/My Drive/Master Thesis/instruction_tuning_rtl.json") as f:
  rtl_data = json.load(f)
```
