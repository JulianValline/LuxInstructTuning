# Clean responses

Since we only need the JSON files, we will exclude the rest of the response

```python
def extract_json(text):
  """Extracts the substring between the first '[' and last ']' characters, including the braces."""
  start = text.find('[')
  end = text.rfind(']')
  if start != -1 and end != -1 and start < end:
    return text[start:end + 1]  # Include the braces
  else:
    return ""  # Return empty string if braces not found or in wrong order
```

```python
# handle invalid JSON
combined_data = []

for json_string in model_responses_clean:
  try:
    data = json.loads(json_string.strip())
    combined_data.extend(data)
  except json.JSONDecodeError:
    print(f"Skipping invalid JSON: {json_string}")
```

# Replace keys

Note: Some files were generated with non-english or invalid keys, such as "Instruktioun", "Répons", "Réponse", "Respon" and "Äntwert". We have to rename these to english key names, to make it uniform with the rest of the file. We use the pop() function.

```python
all_keys = set()
for item in combined_data:
  all_keys.update(item.keys())

all_keys = list(all_keys)
print(all_keys)
```

```python
# replace invalid keys with valid keys
for item in combined_data:
  if "Instruktioun" in item:
    item["instruction"] = item.pop("Instruktioun")
  if "Répons" in item:
    item["response"] = item.pop("Répons")
  if "Äntwert" in item:
    item["response"] = item.pop("Äntwert")
  if "Respon" in item:
    item["response"] = item.pop("Respon")
  if "Réponse" in item:
    item["response"] = item.pop("Réponse")
```

```python
# add new column to json file that combines instruction and response to desired format
# following: https://huggingface.co/datasets/vicgalle/alpaca-gpt4?row=1
for i in range(len(combined_data)):
  instruction = combined_data[i]["instruction"]
  response = combined_data[i]["response"]
  text = f"The following provides an instruction in form of a question. \
  Give a response that answers the question accurately. \
  \n\n###Instruction: {instruction}\n\n###Response: {response}"
  combined_data[i]["text"] = text
combined_data
```

```python
# Write the combined data to a new JSON file
with open('/content/drive/My Drive/Master Thesis/instruction_tuning_wiki.json', 'w') as f:
  json.dump(combined_data, f, indent=4)  # Use indent for pretty printing
```
