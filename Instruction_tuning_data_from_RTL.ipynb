# Create Luxembourgish instruction tuning dataset from RTL news articles

In this notebook, we create an instruction tuning dataset to fine-tune Llama Models from Luxembougish RTL News Data. We use Llama-3.1-405b-instruct to create synthetic data, using a Novita AI API key (https://novita.ai/). We only use 400 samples from the preprocessed Wikipedia dump, to keep the dataset small. We create 5 instruction-answer pairs for each entry, totalling to 2000 synthetic instruction-answer pairs in Luxembourgish.

```python
# define instruction
instruction = "The objective is to create a synthetic dataset in Luxembourgish \
for instruction fine-tuning. You are an assistant and generate all your \
responses in Luxembourgish. You will receive a series of texts one after the \
other. Generate a dataset from the text given with 5 unique pairs of \
instruction and expected response. Give elaborate and high-quality responses \
and answer as accurately as possible. Make sure that the instruction contains \
all relevant information that is needed to give the response. If the question \
does not contain enough information to be answered accurately, reply that you \
need more information to answer the question. Whenever a date is provided for \
a news article, include the date in the instruction and its respective response \
or just in the response, depending on what you consider best in the context. \
Also give temporal context if possible. Name the columns instruction and \
response respectively. Put everything into a JSON file."
instruction
```

```python
from datetime import datetime

def get_date_only(date_string):
  """Returns only the date from a date time string"""
  datetime_object = datetime.strptime(date_string, '%Y-%m-%d %H:%M:%S')
  date = datetime_object.strftime('%Y-%m-%d')
  return date
```

```python
# Check the prompt formatting
for article in rtl_data[:5]:
  title = article.get("title")
  header = article.get("header")
  text = article.get("text")
  date = get_date_only(article.get("public_date"))
  prompt = instruction + "\n\n" + "Text: " + "\n\n" + "News article from: " + date + "\n\n" + title + "\n\n" + header + "\n\n" + text
  print(prompt)
  print("---------------------------------")
```

# Clean responses

Since we only need the JSON files, we will exclude the rest of the response.

```python
def extract_json(text):
  """Extracts the substring between the first '[' and last ']' characters, including the braces."""
  start = text.find('[')
  end = text.rfind(']')
  if start != -1 and end != -1 and start < end:
    return text[start:end + 1]  # Include the braces
  else:
    return ""  # Return empty string if braces not found or in wrong order
```

```python
# handle invalid JSON
combined_data = []

for json_string in model_responses_clean:
  try:
    data = json.loads(json_string.strip())
    combined_data.extend(data)  # append each dictionary to the list
  except json.JSONDecodeError:
    print(f"Skipping invalid JSON: {json_string}")
```

```python
# Write the combined data to a new JSON file
with open('/content/drive/My Drive/Master Thesis/instruction_answer_rtl.json', 'w') as f:
  json.dump(combined_data, f, indent=4)  # Use indent for pretty printing
```

# Replace keys

Note: Some files were generated with non-english or invalid keys, such as "Instruktioun" or "Répons". We have to rename these to english key names, to make it uniform with the rest of the file. We use the pop() function.

```python
all_keys = set()
for item in combined_data:
  all_keys.update(item.keys())

all_keys = list(all_keys)
print(all_keys)
```

```python
# replace invalid keys with valid keys
for item in combined_data:
  if "Instruktioun" in item:
    item["instruction"] = item.pop("Instruktioun")
  if "Répons" in item:
    item["response"] = item.pop("Répons")
  if "Äntwert" in item:
    item["response"] = item.pop("Äntwert")
  if "Respon" in item:
    item["response"] = item.pop("Respon")
  if "Réponse" in item:
    item["response"] = item.pop("Réponse")
```

```python
# add new column to json file that combines instruction and response to desired format
# following: https://huggingface.co/datasets/vicgalle/alpaca-gpt4?row=1
for i in range(len(combined_data)):
  instruction = combined_data[i]["instruction"]
  response = combined_data[i]["response"]
  text = f"The following provides an instruction in form of a question. \
  Give a response that answers the question accurately. \
  \n\n###Instruction: {instruction}\n\n###Response: {response}"
  combined_data[i]["text"] = text
combined_data
```

```python
# Write the combined data to a new JSON file
with open('/content/drive/My Drive/Master Thesis/instruction_tuning_rtl.json', 'w') as f:
  json.dump(combined_data, f, indent=4)  # Use indent for pretty printing
```
